```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv(r"C:\Users\om\Downloads\archive (4)\bengaluru_house_prices.csv")
print(df.shape)
df.head()
```

```python
df2 = df.drop(['area_type', 'society', 'balcony', 'availability'], axis='columns')
df2.isnull().sum()  # Check for remaining missing values
```

```python
# 1. Drop rows with missing values and explicitly make a COPY
df3 = df2.dropna().copy()

# 2. Now you can safely create the 'bhk' column without any warnings!
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))

# 3. Clean the 'total_sqft' column (Handles ranges like '2100 - 2850' by taking average)
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2
    try:
        return float(x)
    except:
        return None

df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4 = df4.dropna() # Drop any rows that couldn't convert properly
```

```python
df4.location = df4.location.apply(lambda x: x.strip())
location_stats = df4.groupby('location')['location'].agg('count').sort_values(ascending=False)

# Any location with less than 10 data points will be labeled 'other'
location_stats_less_than_10 = location_stats[location_stats <= 10]
df4.location = df4.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
```

```python
# 1. Square foot per BHK check (Minimum 300 sqft per bedroom is standard)
df5 = df4[~(df4.total_sqft / df4.bhk < 300)]

# 2. Price per square foot outlier check per location
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price * 100000 / subdf.total_sqft)
        st = np.std(subdf.price * 100000 / subdf.total_sqft)
        reduced_df = subdf[(subdf.price * 100000 / subdf.total_sqft > (m - st)) & (subdf.price * 100000 / subdf.total_sqft <= (m + st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df6 = remove_pps_outliers(df5)

# 3. Bathroom outlier check (It's unusual to have 2 more bathrooms than bedrooms)
df7 = df6[df6.bath < df6.bhk + 2]

# Final features selection
df_final = df7.drop(['size'], axis='columns')
```

```python
# Convert categorical locations into dummy binary columns
dummies = pd.get_dummies(df_final.location).astype(int)
df_ml = pd.concat([df_final.drop('location', axis='columns'), dummies], axis='columns')

# Define features (X) and target variable (y)
X = df_ml.drop('price', axis='columns')
y = df_ml.price

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)

# Check performance accuracy
score = rf_regressor.score(X_test, y_test)
print(f"Model R^2 Accuracy Score: {score * 100:.2f}%")
```

```python
import joblib

# Export model
joblib.dump(rf_regressor, 'house_price_model.pkl')

# Export column structure as a list so Streamlit knows how to arrange inputs
import json
columns = {
    'data_columns' : [col.lower() for col in X.columns]
}
with open("columns.json", "w") as f:
    f.write(json.dumps(columns))
```
